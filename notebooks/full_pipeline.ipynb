# Reproducibility Note

Run this notebook top-to-bottom to reproduce the paper results.
All random seeds are set to 42 via set_all_seeds(). Doc2Vec runs
with workers=1 to fix document processing order.

The pipeline ran three times from scratch and produced
Cohen's kappa = 0.7078 (n=63) identically across all three runs.
These are the results reported in the paper.

⚠️ Important: The LLM validation step (Step 4) calls an external
API (Anthropic Claude, Google Gemini as backup). LLM providers
update their model routing silently between sessions. If you run
this pipeline, your kappa may differ slightly from 0.7078 because
the model version active at your run time may not be identical to
the one used in the archived paper run. This is a known limitation
documented in the paper (Section 5, Limitations).

To get the closest possible reproduction:
  - Pin the model version explicitly in your .env file
  - Use the archived classification outputs in
    outputs/validation/frozen_paper_run/ to skip the LLM step
    and compute kappa directly from the paper's classifications
